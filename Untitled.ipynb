## Import Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("DataCoSupplyChainDataset.csv", encoding='latin1')
df.head()
```

```python
df.info()
```

## Clean Data

```python
# Remove unnecessary personal info
cols_to_drop = ['Customer Email', 'Customer Password', 'Customer Fname', 
                'Customer Lname', 'Customer Street', 'Customer Zipcode']
df.drop(columns=cols_to_drop, inplace=True, errors='ignore')

# Check nulls
print(df.isnull().sum().sort_values(ascending=False).head(10))

# Convert date columns
df['order date (DateOrders)'] = pd.to_datetime(df['order date (DateOrders)'], errors='coerce')
df['shipping date (DateOrders)'] = pd.to_datetime(df['shipping date (DateOrders)'], errors='coerce')

# Derive delay days
df['Delivery_Days'] = (df['shipping date (DateOrders)'] - df['order date (DateOrders)']).dt.days
df['Delay_vs_Scheduled'] = df['Days for shipping (real)'] - df['Days for shipment (scheduled)']
```

## Analysis

```python
# Late deliveries
late = df[df['Late_delivery_risk'] == 1]
print(f"Late deliveries: {len(late)} out of {len(df)}")


```

```python
# Top performing categories
category_sales = df.groupby('Category Name')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,5))
category_sales.head(10).plot(kind='bar', color='skyblue')
plt.title("Top 10 Product Categories by Sales")
plt.ylabel("Total Sales")
plt.show()
```

```python
profit_region = df.groupby('Order Region')['Order Profit Per Order'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
sns.barplot(x=profit_region.values, y=profit_region.index)
plt.title("Profit by Region")
plt.xlabel("Total Profit")
plt.show()

# Correlation
plt.figure(figsize=(8,6))
sns.heatmap(df[['Sales', 'Order Item Discount', 'Order Item Profit Ratio', 
                'Days for shipping (real)']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Compare scheduled vs real delivery days
plt.figure(figsize=(6,4))
sns.histplot(df['Delay_vs_Scheduled'], kde=True, color='orange')
plt.title("Delivery Delay Distribution")
plt.xlabel("Days (Real - Scheduled)")
plt.show()

# Average delay per region
delay_region = df.groupby('Order Region')['Delay_vs_Scheduled'].mean().sort_values()
delay_region.plot(kind='barh', color='teal', figsize=(8,4), title='Avg Delay by Region')
plt.show()
```

```python
cleaned_df = df[['Order Id', 'Order Region', 'Category Name', 'Sales', 
                 'Order Profit Per Order', 'Order Item Discount', 
                 'Order Item Profit Ratio', 'Late_delivery_risk', 
                 'Delay_vs_Scheduled', 'Shipping Mode', 
                 'order date (DateOrders)', 'shipping date (DateOrders)']]

cleaned_df.to_csv("Cleaned_SupplyChain.csv", index=False)
print("Cleaned dataset exported for Power BI visualization!")
```
